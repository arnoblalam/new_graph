```python
%pylab inline
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
import tree
from collections import OrderedDict
```

```python
royal_family = OrderedDict([
    ("Elizabeth II", {"Charles", "Andrew", "Edward", "Anne"}),
    ("Charles", {"William", "Henry"}),
    ("William", {"George", "Charlotte"}),
    ("George", set()),
    ("Charlotte", set()),
    ("Henry", set()),
    ("Andrew", {"Beatrice", "Eugenie"}),
    ("Beatrice", set()),
    ("Eugenie", set()),
    ("Edward", {"Severn", "Louise"}),
    ("Severn", set()),
    ("Louise", set()),
    ("Anne", {"Peter", "Zara"}),
    ("Peter", {"Savannah", "Isla"}),
    ("Savannah", set()),
    ("Isla", set()),
    ("Zara", {"Mia"}),
    ("Mia", set())])

royal_family_weights = {
    "Elizabeth II": 1024,
    "Charles": 512,
    "William": 256,
    "George": 128,
    "Charlotte": 64,
    "Henry": 32,
    "Andrew": 16,
    "Beatrice": 8,
    "Eugenie": 4,
    "Edward": 2,
    "Severn": 1,
    "Louise": 0.5,
    "Anne": 0.25,
    "Peter": 0.125,
    "Savannah": 0.0625,
    "Isla": 0.03125,
    "Zara": 0.015625,
    "Mia": 0.0078125
}
```

```python
tree.draw_tree(royal_family)
```

```python
results = tree.aggregate(t=royal_family, node_weights=royal_family_weights, desired_level=2, how_many=50)
```

```python
results
```

```python
results.plot(x="nodes", y = "normalized_entropy")
```
